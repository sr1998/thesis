### Pre-processing the data

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Normalizer

columns_to_keep = ["Sample", "Group", "Project", "Project_1"]

def centered_arcsine_transform(df: pd.DataFrame) -> pd.DataFrame:
    """Centered arcsine transform of a dataframe.

    Assumes that the dataframe contains proportions of some kind, i.e. values in [0, 1].
    """
    def arcsine(x):
        return np.arcsin(np.sqrt(x))
    
    r, c = df.shape
    C = np.eye(c) - 1/c * np.ones((c, c))
    df_vals = df.to_numpy()
    df_vals = arcsine(df_vals)
    df_vals = df_vals @ C
    df = pd.DataFrame(df_vals, index=df.index, columns=df.columns)
    return df
    

# testing centered arcsine transform
df = pd.DataFrame(np.random.rand(10, 5), columns=[f'col_{i}' for i in range(5)])
df = pd.DataFrame(Normalizer('l1').fit_transform(df), index=df.index, columns=df.columns)
df_copy = df.copy()
display(df)
df = centered_arcsine_transform(df)
display(df)

# testing whether splitting the data into two parts (row-wise) and then applying the transform gives the same result as applying the transform to the whole data
df1 = df_copy.iloc[:3, :]
df2 = df_copy.iloc[3:, :]

df1 = centered_arcsine_transform(df1)
df2 = centered_arcsine_transform(df2)

df_combined = pd.concat([df1, df2])
display(df_combined)

print(df.equals(df_combined)) # should be True
```
